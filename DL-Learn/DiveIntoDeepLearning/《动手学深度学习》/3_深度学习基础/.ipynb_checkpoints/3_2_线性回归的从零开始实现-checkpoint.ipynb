利用 `Tensor` 和 `autograd` 来实现一个线性回归的训练

导入包

```python
%matplotlib inline
import torch
from IPython import display
from matplotlib import pyplot as plt
import numpy as np
import random
import sys
sys.path.append("..")
import d2lzh_pytorch
```

### 3.2.1 生成数据集

构造人工训练数据集。

样本数1000，特征数2，权值w=[2,-3.4]，偏差b=4.2，随机噪声e（服从均值为0，标准差为0.01的正态分布）

y = Xw + b + e

```python
num_inputs = 2
num_examples = 1000
true_w = [2, -3.4]
true_b = 4.2
features = torch.randn(num_examples, num_inputs, dtype=torch.float32)
labels = true_w[0] * features[:, 0] + true_w[1] * features[: , 1] + true_b
labels += torch.tensor(np.random.normal(0, 0.01, size=labels.size()),
                      dtype=torch.float32)
```

```python
print(features[0], labels[0])
```

```python
# 生成散点图，直观地观察两者间的线性关系
# 展示
d2lzh_pytorch.set_figsize()
plt.scatter(features[:, 1].numpy(), labels.numpy(), 1)
```

### 3.2.2 读取数据

训练模型时，需要遍历数据集并不断读取小批量数据样本

```python
def data_iter(batch_size, features, labels):
    num_examples = len(features)
    indices = list(range(num_examples))
    random.shuffle(indices)  # 打乱样本顺序
    for i in range(0, num_examples, batch_size):
        j = torch.LongTensor(indices[i: min(i + batch_size, num_examples)])
        yield features.index_select(0, j), labels.index_select(0, j)  # 0表示按行索引
```

```python
batch_size = 10

for X, y in data_iter(batch_size, features, labels):
    print(X, y)
    break
```

### 3.2.3 初始化模型参数

权重初始化为均值为0，标准差为0.01的正态随机数，偏差初始化为0

```python
w = torch.tensor(np.random.normal(0, 0.01, (num_inputs, 1)), dtype=torch.float32)
b = torch.zeros(1, dtype=torch.float32)
```

模型训练中，需要对参数求梯度来迭代参数的值

```python
w.requires_grad_(requires_grad=True)
b.requires_grad_(requires_grad=True)
```

### 3.2.4 定义模型

```python
def linreg(X, w, b):
    return torch.mm(X, w) + b
```

### 3.2.5 定义损失函数

```python
def squared_loss(y_hat, y):
    return (y_hat - y.view(y_hat.size())) ** 2 / 2
```

### 3.2.6 定义优化算法

小批量随机梯度下降算法

```python
def sgd(params, lr, batch_size):
    for param in params:
        param.data -= lr * param.grad / batch_size
```

### 3.2.7 训练模型

训练中，多次迭代模型参数。

每次迭代中，根据当前读取的小批量数据样本（特征 X 和标签 y），通过调用反向函数 `backward` 计算小批量随机梯度，并调用优化算法 `sgd` 迭代模型参数。

**由于 batch_size 为10，所以每个小批量的损失 `l` 的形状为（10，1）**

**变量`l`并不是一个标量，所以调用 `.sum()` 将其求和得到一个标量，再运行 `l.backward()` 得到该变量有关模型参数的梯度**

**注意：每次更新完参数后，需要将参数的梯度清零**

每个迭代周期，将完整遍历一遍 `data_iter` 函数。

迭代周期 `num_epoches` 和 学习率 `lr` 都是超参数。

```python
lr = 0.03  # 学习率
num_epoches = 3  # 迭代周期
net = linreg  # 网络模型
loss = squared_loss  # 损失函数

for epoch in range(num_epoches):
    # 每个迭代周期，是使用训练集的所有样本一次
    # X和y是小批量样本的特征和标签
    for X, y in data_iter(batch_size, features, labels):
        l = loss(net(X, w, b), y).sum()
        l.backward()
        sgd([w, b], lr, batch_size)
        
        w.grad.data.zero_()
        b.grad.data.zero_()
    train_l = loss(net(features, w, b), labels)
    print('epoch %d, loss %f' % (epoch + 1, train_l.mean().item()))
```

```python
# 比较学习到的参数和真实参数
print(true_w, w)
print(true_b, b)
```
